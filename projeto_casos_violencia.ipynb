```python
import pandas as pd
import plotly.express as px
from  dash_bootstrap_templates  import  ThemeSwitchAIO
pd.options.display.max_rows = 1000
pd.options.display.max_columns = 220


dados = pd.read_csv('dados_violencia.csv')


tabela_filtrada = dados[['DT_NOTIFIC','NU_ANO','DT_OCOR','idade.anos','CS_SEXO',
                         'CS_GESTANT','CS_RACA','CS_ESCOL_N',
                         'SIT_CONJUG','LOCAL_OCOR','AG_ESPEC','SEX_ASSEDI','SEX_ESTUPR','PEN_ORAL','PEN_ANAL','PEN_VAGINA'
                         ,'CONS_SUIC','AUTOR_SEXO','ENC_IML','ID_PAIS']]

dados.head(2)
```

```python
tabela_violencia = dados[['VIOL_FISIC','VIOL_PSICO','VIOL_TORT','VIOL_SEXU','VIOL_TRAF','VIOL_FINAN','VIOL_NEGLI'
                      ,'VIOL_INFAN','VIOL_LEGAL','VIOL_OUTR']]

novos_nomes_colunas = ['FISIC','PSICO','TORT','SEXU','TRAF','FINAN','NEGLI'
                      ,'INFAN','LEGAL','OUTR']

tabela_AG = dados[['AG_FORCA','AG_ENFOR','AG_OBJETO','AG_CORTE','AG_QUENTE','AG_ENVEN','AG_FOGO'
                      ,'AG_AMEACA','AG_OUTROS']]

novos_nomes_colunas2 = ['FORCA','ENFOR','OBJETO','CORTE','QUENTE','ENVEN','FOGO'
                      ,'AMEACA','OUTROS']

tabela_sex = dados[['SEX_ASSEDI','SEX_ESTUPR','SEX_PUDOR','SEX_PORNO','SEX_EXPLO','SEX_OUTRO','PEN_ORAL','PEN_ANAL','PEN_VAGINA',]]

novos_nomes_colunas3 = ['ASSEDI','ESTUPR','PUDOR','PORNO','EXPLO','OUTRO','ORAL','ANAL','VAGINA',]

tabela_sex.columns = novos_nomes_colunas3

tabela_AG.columns = novos_nomes_colunas2

# Atribua os novos nomes às colunas do DataFrame
tabela_violencia.columns = novos_nomes_colunas

tabela_sex
```

```python
def obter_colunas_sim(row,tabela):
    colunas_sim = [coluna for coluna in tabela.columns if row[coluna] == 'sim']
    return ', '.join(colunas_sim) if colunas_sim else 'Nenhuma'

# Aplique a função a cada linha do DataFrame e crie uma nova coluna chamada 'colunas_sim'
tabela_violencia['Tipo_Violencia'] = tabela_violencia.apply(lambda row: obter_colunas_sim(row,tabela_violencia), axis=1)

tabela_AG['Tipo_agreçao'] = tabela_AG.apply(lambda row: obter_colunas_sim(row,tabela_AG), axis=1)

tabela_sex['tipo_crime_sex'] = tabela_sex.apply(lambda row: obter_colunas_sim(row,tabela_sex), axis=1)
# Exiba o DataFrame resultante
tabela_violencia
```

```python
tabela_filtrada['Data'] = pd.to_datetime(tabela_filtrada['DT_NOTIFIC'])

tabela_filtrada_data  =  tabela_filtrada.set_index('Data')

tabela_filtrada_data.rename(columns={'ID_PAIS': 'Casos'}, inplace=True)

serie_temporal = tabela_filtrada_data['Casos']

data_ano = serie_temporal.resample('A').sum()

data_mes = serie_temporal.resample('M').sum()
```

```python
fig = px.area(data_ano,template='cyborg',labels={'Data':'','value':'Total de denuncias','variable':'numero de'},
                    title='progressao de denuncias anual',height=500,width=1600,range_x=['2011-01-01','2020-12-31'])
fig.update_layout(font=dict(size=25),title_font=dict(size=25),showlegend=False)
fig.update_xaxes(dtick='M12',)
fig.show()
```

```python
fig2 = px.area(data_mes,template='cyborg',labels={'Data':'','value':'Total de denuncias','variable':'numero de'},
                    title='progressao de denuncias semestral',height=500,width=1600,range_x=['2010-01-01','2020-12-31'])
fig2.update_layout(font=dict(size=25),title_font=dict(size=25),showlegend=False)
fig2.update_xaxes(dtick='M6',)
fig2.show()
```

```python
def grafico_info(base_dados,titulo):
    filtrando_os_sim3 = base_dados[base_dados == 'sim'].count().sort_values()
    transformando_df3 = pd.DataFrame(filtrando_os_sim3,columns=['A'])
    df_ferramenta = transformando_df3.reset_index()
    
    fig8 = px.histogram(df_ferramenta,x='index',y='A',template='cyborg',text_auto=True,
                    title=titulo,height=500,width=1550)
    fig8.update_layout(font=dict(size=25),title_font=dict(size=20),showlegend=False,yaxis_title=None,xaxis_title=None)
    fig8.show()

def grafico_info2(coluna,titulo):
    fig5 = px.histogram(sex_vitima_agressor,x= coluna ,template='cyborg',text_auto=True,
                    title=titulo ,height=500,width=1600,)
    fig5.update_layout(font=dict(size=25),title_font=dict(size=25),showlegend=False,yaxis_title=None,xaxis_title=None)
    fig5.show()
```

```python
seleçao = sex_vitima_agressor.value_counts()
seleçao.reset_index()
df_vitma_agressor = pd.DataFrame(seleçao)
df_user = df_vitma_agressor.reset_index()

fig6 = px.histogram(df_user,x='CS_SEXO',y=0,template='cyborg',text_auto=True,color='AUTOR_SEXO',barmode='group',
                    title='vitima e autor do crime',height=500,width=1550,)
fig6.update_layout(font=dict(size=25),title_font=dict(size=25),yaxis_title=None,xaxis_title='sexo vitima')
fig6.show()
```

```python
def teste1(tabela,ano_ou_mes):
    def teste2(x):
        if x == 'sim':
            return 1
        else:
            return 0
        
    serie_temporal2 = tabela_filtrada_data[tabela]

    serie = serie_temporal2.apply(teste2)

    data_ano2 = serie.resample(ano_ou_mes).sum()

    fig = px.line(data_ano2,template='cyborg',labels={'Data':'','value':'Total de denuncias','variable':'numero de'},
                    title='progressao de denuncias anual',height=500,width=1600,range_x=['2010-01-01','2020-12-31'])
    
    fig.update_layout(font=dict(size=25),title_font=dict(size=25),showlegend=False)

    fig.update_xaxes(dtick='M12')
    
    return fig.show()
```

```python
def teste33(tabela,ano_ou_mes):
    def teste23(x):
        if x == 'sim':
            return 1
        else:
            return 0
        
    serie_temporal2 = tabela_filtrada_data[tabela]

    serie = serie_temporal2.apply(teste23)

    data_ano2 = serie.resample(ano_ou_mes).sum()
    return data_ano2
```
